```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

<b>Task:</b> Use the rest of this notebook to carry out your project plan. 

You will:

1. Prepare your data for your model.
2. Fit your model to the training data and evaluate your model.
3. Improve your model's performance by performing model selection and/or feature selection techniques to find best model for your problem.

Add code cells below and populate the notebook with commentary, code, analyses, results, and figures as you see fit. 

```python
# YOUR CODE HERE
#Features and target like we mentioned
features = ['neighbourhood_group_cleansed', 'price', 'minimum_nights', 'number_of_reviews', 'availability_365', 'reviews_per_month']
target = 'room_type'

X = df[features].copy()
y = df[target]

# Filling values with mean value for numerical features so as to handle missing values
X['reviews_per_month'].fillna(X['reviews_per_month'].mean(), inplace=True)
X['number_of_reviews'].fillna(0, inplace=True)  # If no reviews, assume 0

# Encoding categorical features and scale numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), ['price', 'minimum_nights', 'number_of_reviews', 'availability_365', 'reviews_per_month']),
        ('cat', OneHotEncoder(), ['neighbourhood_group_cleansed'])
    ])

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pipeline_lr = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', LogisticRegression(max_iter=1000))])

# Training model
pipeline_lr.fit(X_train, y_train)

# Prediction on test set
y_pred_lr = pipeline_lr.predict(X_test)

# Evaluating the model
print("Logistic Regression:")
print(classification_report(y_test, y_pred_lr))
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
```

```python
pipeline_rf = Pipeline(steps=[('preprocessor', preprocessor),
                              ('classifier', RandomForestClassifier())])

# Training model
pipeline_rf.fit(X_train, y_train)

# Predicting on test set
y_pred_rf = pipeline_rf.predict(X_test)

# Evaluateing the model
print("Random Forest:")
print(classification_report(y_test, y_pred_rf))
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
```

```python
# Hyperparameter tuning for Random Forest
param_grid = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [10, 20, 30]
}

# Grid search
grid_search = GridSearchCV(estimator=pipeline_rf, param_grid=param_grid, cv=3, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
y_pred_rf_optimized = grid_search.predict(X_test)

print("Optimized Random Forest:")
print(classification_report(y_test, y_pred_rf_optimized))
print("Accuracy:", accuracy_score(y_test, y_pred_rf_optimized))
```
